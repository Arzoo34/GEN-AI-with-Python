**LangChain Expression Language**


*   It is a declarative way to easily chain LangChain components together.
*   It was designed from the ground up to support bringing chains from prototype to production with no code changes, handling everything from streaming to parallel execution.

**The Pipe Operator ( | )**


*   LCEL relies heavily on the pipe operator, which works just like a unix pipe. It takes the output of the component on the left and passes it as input to the componet on the right.
*   chain = prompt | model | output_parser

**KEY FEATURES**


*   First-Class Streaming Support: you can directly stream token from the model, even if your chain has multiple steps. Instead of invoke() you just call stream()
      
*   LCEL automatically executes methods in parallel using RunnableParallel.


*   Built-in Batching : you can run a batch of inputs through your chain using .batch()

**responses = chain.batch([{"topic": "bears"}, {"topic": "cats"}])**

```python
import os
os.environ["GOOGLE_API_KEY"] = "api"
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
model = ChatGoogleGenerativeAI(model = "gemini-2.5-flash",temperature = 0.7)
prompt = ChatPromptTemplate.from_template("Tell me a brief introduction about {topic}")
output_parser = StrOutputParser()
chain =prompt|model|output_parser
response = chain.invoke({"topic":"quantum computing"})
print(response)
```

# **chain = prompt | model | output_parser**


### 1.   **THE INPUT(The Starting Dictionary):**

{topic: "quantum computing"}

### 2.  **prompt (ChatPromptTemplate):**

a. A structured template ("Tell me a brief, interesting fact about {topic}").


b. **Its Role:** It acts as a Data Formatter. It catches the raw input dictionary from Step 1, extracts the value of {topic}, and injects it into the template.

c.** Output:** It outputs a formal PromptValue object containing a list of structured chat messages (e.g., [HumanMessage(content="Tell me a brief, interesting fact about quantum computing")]).

### 3. **model (ChatGoogleGenerativeAI):**
 a. What it is: The wrapper for Google Gemini.

 b. **Its Role:** It acts as the Engine / Brain. It receives the structured PromptValue from the prompt component, sends it over the internet via API to Google's servers, and tells Gemini (gemini-2.5-flash) to process it.

 c.** Output:** It outputs an AIMessage object. This isn't raw text yet; it's a rich metadata object containing the text response along with usage statistics, finish reasons, and safety ratings.

### 4. **output_parser (StrOutputParser)**
What it is: A utility that cleans up raw model responses.

**Its Role:** It acts as the Filter / Extraction layer. Because the model outputs a complex AIMessage object, this parser isolates just the text generation inside that object and strips away the unnecessary metadata.

##### **Output:** A clean, standard Python string (str) containing just the answer, which is what finally gets printed to your console.
